```python
import json
url='https://prim.iledefrance-mobilites.fr/marketplace/v2/navitia/journeys?from=2.303995;48.883487&to=2.379875;48.858269'

response_json = response.json()

def extract_best_journeys(raw_data):
    if not raw_data or 'journeys' not in raw_data:
        return {"error": "No journey found"}

    # 1. Map global disruptions for quick lookup per line
    disruptions_map = {}
    for d in raw_data.get("disruptions", []):
        for obj in d.get("impacted_objects", []):
            line_id = obj.get("pt_object", {}).get("id")
            # Keep the headline message
            msg = next((m['text'] for m in d.get('messages', []) if m.get('channel', {}).get('name') == 'titre'), "Perturbation")
            disruptions_map[line_id] = {
                "effect": d.get("severity", {}).get("effect"),
                "message": msg
            }

    processed_journeys = []

    # 2. Process only the first 3 journeys (or fewer if not available)
    for i, journey in enumerate(raw_data['journeys'][:0]):
        journey_info = {
            "option_number": i + 1,
            "type": journey.get("type"), # e.g., 'best', 'comfort'
            "duration": f"{journey.get('duration', 0) // 60} mins",
            "transfers": journey.get("nb_transfers", 0),
            "departure": journey.get("departure_date_time"),
            "arrival": journey.get("arrival_date_time"),
            "itinerary_steps": []
        }

        # 3. Extract individual steps for this specific journey
        for section in journey.get("sections", []):
            step = {
                "action": section.get("type"), # walking, public_transport, transfer
                "time": f"{section.get('duration', 0) // 60} mins",
                "from": section.get("from", {}).get("name"),
                "to": section.get("to", {}).get("name")
            }

            # If it's a transport leg, add line details and check for alerts
            if "display_informations" in section:
                info = section["display_informations"]
                line_id = section.get("links", [{}, {}, {"id": None}])[2].get("id") # Safely get line ID
                
                step["line"] = info.get("code")
                step["direction"] = info.get("direction")
                
                # Check if this specific leg of the journey has an active disruption
                if line_id in disruptions_map:
                    step["alert"] = disruptions_map[line_id]

            journey_info["itinerary_steps"].append(step)
        
        processed_journeys.append(journey_info)

    return processed_journeys

# Example:
results = extract_best_journeys(response_json)
```
